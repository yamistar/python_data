# 1. 데이터읽기

```python
# 로켓발사 데이터를 읽어들여 변수에 할당
launch_data = pd.read_csv('../data/RocketLaunchDataCSV.csv')
launch_data.head()
```

# 3. 데이터 전처리

1. 결측치 처리
2. 범주형  데이터 인코딩
3. 특징 스케일링

```python
# 누락값을 정리하기 위해서 적당한 값으로 누락값을 처리합니다.
launch_data['Launched?'].fillna('N', inplace=True)
launch_data['Crewed or Uncrewed'].fillna('Uncrewed', inplace=True)
launch_data['Wind Direction'].fillna('Unknown', inplace=True)
launch_data['Condition'].fillna('Fair', inplace=True)
launch_data.fillna(0, inplace=True)
launch_data.head()
```

```python
from sklearn import preprocessing
```

```python
# 머신러닝 모델의 입력으로 숫자 데이터를 사용해야 하므로 모든 텍스트를 숫자로 변환한다. 예를 들어 유인 로켓이면 1, 무인로켓이면 0
label_encoder = preprocessing.LabelEncoder()

# 카테고리를 나타내는 텍스트를 가진 아래 3개의 컬럼을 숫자 데이터를 가지도록 변환한다.
launch_data['Crewed or Uncrewed'] = label_encoder.fit_transform(launch_data['Crewed or Uncrewed'])
launch_data['Wind Direction'] = label_encoder.fit_transform(launch_data['Wind Direction'])
launch_data['Condition'] = label_encoder.fit_transform(launch_data['Condition'])
```

```python
# 머신러닝 모델의 출력으로 우리가 예측하고 싶은 것은 'Launched?'컬럼이다. 이것을 출력 변수에 할당한다.
y = launch_data['Launched?']
```

```python
launch_data.columns
```

```python
# 이미 실행이 된 다음에는 KeyError 발생함.

# 머신러닝 모델의 입력값으로 컬럼을 선택한다. 선택한 컬럼들이 feature이다.
launch_data.drop(['Name', 'Date', 'Time (East Coast)', 'Location', 'Launched?', 'Hist Ave Sea Level Pressure', 'Sea Level Pressure', 'Day Length', 'Notes', \
                  'Hist Ave Visibility', 'Hist Ave Max Wind Speed'], axis=1, inplace=True)
X = launch_data
```

```python
X.info()
```

# 4. 모델 학습

## 의사결정나무

```python
# Machine learning libraries used to build a decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# 의사 결정 나무 분류자를 생성한다.
tree_model = DecisionTreeClassifier(random_state=0, max_depth=5)
tree_model
```

```python
# SKlearn libary contains all the machine learning packages we need to digest and extract patterns from the data
from sklearn import linear_model, model_selection, metrics
from sklearn.model_selection import train_test_split
```

```python
# 데이터를 학습 데이터와 테스트 데이터로 분할한다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
# 학습 데이터로 머신러닝 모델 학습 시키기
tree_model.fit(X_train, y_train)
```

```python
# 머신러닝 모델 테스트
y_pred = tree_model.predict(X_test)
y_pred
```

```python
# 머신러닝 모델 평가
tree_model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 혼동 행렬
cm1 = confusion_matrix(y_test, y_pred)

# 혼동 행렬 시각화
plt.figure(figsize=(10, 7))
sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pydotplus
from IPython.display import Image
```

```python
# 시각화를 위한 함수를 작성한다.
def tree_graph_to_png(tree, feature_names, class_names, png_file_to_save):
    tree_str = export_graphviz(tree, feature_names=feature_names, class_names=class_names, filled=True, out_file=None)
    graph = pydotplus.graph_from_dot_data(tree_str)
    print(graph)
    graph.write_png(png_file_to_save)
    return Image(graph.create_png())
```

```python
# 함수를 호출하여 시각화 한다.
tree_graph_to_png(tree=tree_model, feature_names=X.columns.values, class_names=['No Launch', 'Launch'], png_file_to_save='decision-tree.png')
```
