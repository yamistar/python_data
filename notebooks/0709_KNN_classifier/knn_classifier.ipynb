```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 기존 데이터 닥스훈트와 진돗개
dach_length = [55, 57, 64, 63, 58, 49, 54, 61]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34]

jin_length = [56, 47, 56, 46, 49, 53, 52, 48]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54]

# 새로운 데이터
newdata_length = [59]
newdata_height = [35]
```

```python
plt.scatter(dach_length, dach_height, label ='Dachshund')
plt.scatter(jin_length, jin_height, label ='Jindo Dog')
# new data
plt.scatter(newdata_length, newdata_height, marker='p')
plt.legend(loc='upper right')
plt.show()
```

```python
# 데이터 테이블로 만들기
d_data = np.column_stack((dach_length, dach_height))
d_label = np.zeros(len(d_data))     # 라벨만들기

j_data = np.column_stack((jin_length, jin_height))
j_label = np.ones(len(j_data))      # 라벨만들기

# 데이터 결합
dogs = np.concatenate((d_data, j_data))
labels = np.concatenate((d_label, j_label))

dogs, labels
```

```python
# 분류기 생성
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn
```

```python
# 학습시키기
knn.fit(dogs, labels)
```

```python
# 새로운 데이타(newdata_)도 합쳐준다.
newdata = np.column_stack((newdata_length, newdata_height))
```

```python
# predict는 모델이 학습한 패턴을 바탕으로 새로운 입력 데이터에 대한 출력 값을 예측하는 메소드
y_pred = knn.predict(newdata)
y_pred
```

```python

dog_classes = {0:'닥스훈트', 1:'진도개'}
print(f'예측데이터 : {newdata} , 예측결과 : {dog_classes[y_pred[0]]}')
```

```python
# 진도개 예측 테스트

newdata1 = np.array([[50,41]])
newdata1
```

```python
# 분류기 생성
for k in [1,3,5,9] :
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(dogs, labels)
    y_pred1 = knn.predict(newdata1)
    print(f'k={k} 일 때 - 예측데이터 : {newdata1} , 예측결과 : {dog_classes[y_pred1[0]]}')
```

```python
plt.scatter(dach_length, dach_height, label ='Dachshund')
plt.scatter(jin_length, jin_height, label ='Jindo Dog')
# new data
plt.scatter(newdata1[:,0],newdata1[:,1], marker='p')
plt.legend(loc='upper right')
plt.show()
```
