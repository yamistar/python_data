```python
from sklearn.cluster import KMeans
import numpy as np

# 난수 생성을 위한 seed를 설정 
np.random.seed(42)

# 세 개의 데이터 세트를 생성  각각의 데이터 세트는 평균이 [2,2], [-2,-2], [2,-2]인 정규 분포를 따름
data1 = np.random.randn(100,2) + [2,2]
data2 = np.random.randn(100,2) + [-2,-2]
data3 = np.random.randn(100,2) + [2,2]

# 세 개의 데이터 세트를 하나로 합침
X = np.vstack((data1, data2, data3))
```

```python
# 데이터를 산점도로 그림
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1]) 
```

```python
# KMeans 클러스터링 알고리즘을 초기화  클러스터의 수는 2개로 설정함
kmeans = KMeans(n_clusters=2, random_state=0) 

# KMeans 알고리즘을 데이터에 적합시킴
kmeans.fit(X) 
```

```python
# 예측한 결과
# 각 데이터 포인트가 속한 클러스터의 레이블을 출력 
print(kmeans.labels_)

# 중심점
# 클러스터의 중심점을 계산 
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# 데이터와 클러스터의 중심점을 산점도로 그림
plt.scatter(X[:,0],X[:,1])
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='x', s=200, label='centroids') 

plt.show() 
```

```python
# 새로운 데이터 포인트에 대해 클러스터를 예측 
new_data = np.array([[0, 0], [3, 3], [-3, -3], [2, -3]])
new_labels = kmeans.predict(new_data)
new_labels
```

```python
# 데이터를 산점도로 그림. 각 데이터 포인트의 색상은 해당 데이터 포인트가 속한 클러스터를 나타 냄
# plt.scatter(X[:,0], X[:,1], c=labels) 
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)

# 클러스터의 중심점을 산점도에 표시  중심점은 빨간색 'x'로 표시됨
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='x', s=200, label='centroids') 

# 새로운 데이터 포인트를 산점도에 표시  새로운 데이터 포인트는 삼각형으로 표시되며, 색상은 해당 데이터 포인트가 속한 클러스터를 나타 냄
plt.scatter(new_data[:,0], new_data[:,1], s=200,  c=new_labels, marker='^', label='new data') 

plt.show() 
```

# DBSCAN

```python
# make_blobs 함수 클러스터링을 위한 테스트 데이터를 생성
from sklearn.datasets import make_blobs  # 부족한 데이터 생성용 패키지 

# 클러스터의 중심점을 설정 데이터 생성을 위한 것이며, DBSCAN에 직접적으로 설정하는 것은 아님
centers = [[1, 1], [-1, -1], [1, -1]]

# make_blobs 함수를 사용하여 테스트 데이터를 생성
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)
X.shape, labels_true.shape
```

```python
# 데이터를 표준화
from sklearn.preprocessing import StandardScaler 

X_scaled = StandardScaler().fit_transform(X) 
X_scaled 
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics

# DBSCAN 모델 정의 및 학습하기 
db = DBSCAN(eps=0.3, min_samples=10) 
# eps는 클러스터의 밀도를 결정하는 거리 값, min_samples는 클러스터를 형성하기 위해 필요한 최소 데이터 포인트 수.
labels_db = db.fit_predict(X) 
labels_db[:100] 
```

```python
# 생성된 클러스터의 종류(집합) 확인
unique_labels = set(labels_db) 
```

```python
# 노이즈의 개수 확인
list(labels_db).count(-1) 

# 위와 같은 코드 labels_db[labels_db == -1].size 
```

```python
# 결과 시각화 
# 색상 (스펙트럼) 설정 
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))] 

for k, col in zip(unique_labels, colors): 
    if k == -1: 
        # 노이즈는 검은색으로 표시 
        col = [0, 0, 0, 1] 

    # 현재 클러스터(k)에 속하는 데이터 마스크 설정 
    class_member_mask = (labels_db == k) 

    # 현재 클러스터(k)에 속하는 데이터 포인트 추출 
    xy = X_scaled[class_member_mask] 
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6) 

plt.title('DBSCAN Clustering') 
plt.xlabel("X axis") 
plt.ylabel("Y axis") 
plt.show() 
```

# DBSCAN  튜닝하기

```python
# 튜닝튜닝하기.........

from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
    )
X_scaled = StandardScaler().fit_transform(X) # 표준화

# DBSCAN 모델 정의 eps 값을 수정하기
db = DBSCAN(eps=0.2, min_samples=10)
labels_db = db.fit_predict(X_scaled) # 학습
unique_labels = set(labels_db)

# 결과 시각화 
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]
    class_member_mask = (labels_db == k)
    xy = X_scaled[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)

plt.title('DBSCAN Clustering')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
