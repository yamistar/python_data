```python
# 
import bikes_data
import seaborn as sns
import matplotlib as plt

bikes = bikes_data.data_preprocessing() 
bikes.head(3)
```

# 1. 시간대 x 대여구 데이터 분석

```python
# 시간 장소 특징

hourly_gu_use = bikes.pivot_table(index='대여시간대',columns='대여구',values='자전거번호',aggfunc='count')
hourly_gu_use.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# hourly_gu_use.plot()

plt.figure(figsize=(10,8))
sns.heatmap(data=hourly_gu_use.T)

plt.figure(figsize=(15,8))
sns.boxplot(hourly_gu_use) 

# sns.boxenplot
```

# 2. # 1. 일자 x 대여구 데이터 분석

```python
# 2. # 1. 일자 x 대여구 데이터 분석

daily_gu_use = bikes.pivot_table(index='일자',columns='대여구',values='자전거번호',aggfunc='count')
daily_gu_use.head()
```

```python
# plt.figure(figsize=(10,8))
# sns.heatmap(date.T)

# plt.figure(figsize=(15,8))
# sns.boxenplot(date)


# Create a heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(data=daily_gu_use.T)

# Create a boxplot
plt.figure(figsize=(15, 8))
sns.boxplot(data=daily_gu_use)
```
