## 산점도 그래프
- X축 Y축 데이터 관계 파악

```python

data = pd.read_excel(r"data\health_screenings_2020_1000ea.xlsx")
data
```

```python
data.info()
#  4   height         1000 non-null   int64         
#  5   weight         1000 non-null   int64   
# 키와몸무게의 상관관계 알고 싶다
```

```python
heightdata = data['height']
weightdata = data['weight']

plt.figure(figsize=(15,4))
plt.scatter(heightdata, weightdata)
```

```python
# 수축기 혈합, 이완기 혈합, 공복 혈당 그룹 산점도 그래프

systolic_data = data['systolic']
diastolic_data = data['diastolic']
blood_sugar_data = data['blood_sugar']

plt.figure(figsize=(15,6))

plt.scatter(systolic_data, diastolic_data, color='r', label='systolic * diastolic')
plt.scatter(systolic_data, blood_sugar_data, color='g', label='systolic * blood_sugar')
plt.scatter(diastolic_data, blood_sugar_data, color='b', label='diastolic * blood_sugar')

plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 Health Screenings Group Scatter Graph')
plt.legend()
```

```python
# 여성의 성별, 키 데이터로 여성 키 히스토 그램 

# 데이터셋 생성
# female_data = data.loc[data['gender']==2][['gender','height']]
female_data = data.loc[data['gender']==2, ['gender', 'height']] 

# female_data
plt.figure(figsize=(10,6))
plt.hist(female_data['height'], bins=7, label='Female')

plt.xlim(130,180)
plt.ylim(30,300)
plt.title('Histogram with female gender, height data')
plt.legend()
plt.grid()
plt.show()
```

```python
# 그룹 히스토그램 작성
# 여성, 남성 키 히스토그램

# 데이터셋 생성
female_data = data.loc[data['gender']==2, ['gender', 'height']] 
male_data = data.loc[data['gender']==1, ['gender', 'height']] 

# female_data
plt.figure(figsize=(10,6))
plt.hist(female_data['height'], alpha = 0.5, bins=7, label='Female')
plt.hist(male_data['height'],  alpha = 0.5,bins=7, label='Male')

plt.xlim(130,199)
# plt.ylim(30,300)
plt.title('Histogram with female gender, height data')
plt.legend()
plt.grid()
plt.show()
```

```python
# 상자수염 그래프 작성
import numpy as np

height = np.array([ male_data['height'], female_data['height'] ], dtype=object)

plt.figure(figsize=(10,6))
plt.boxplot(height, label=['Male','Female'])

plt.ylim(130,200)
plt.title('boxplot with gender, height data')
plt.show()
```
